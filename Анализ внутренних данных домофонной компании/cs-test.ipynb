# Описание задания

Структура `data.xslx`:
* `ИД дома` - идентификатор дома
* `ИД подъезда` - идентификатор подъезда в доме (может быть несколько подъездов в 1 доме)
* `ИД квартиры` - идентификатор квартиры в подъезде (может быть несколько квартир в 1 подъезде)
* `Лицевой счет` (ЛС) - номер лицевого счета в квартире (может быть несколько в 1 квартире)
* `Услуга` - услуга периодического обслуживания (может быть несколько услуг на 1 ЛС) 
* `Тариф` - ставка тарифа в месяц по услуге, действующая на момент выгрузки данных.
* `Статус услуги` - название статуса услуги. 
> * Одновременно на одной квартире может быть только одна базовая услуга ТО МКД, ТО СВОД или ТО ДВЕРЬ в статусе! = Завершена. 
> * Остальные услуги дополнительные и могут быть на одном ЛС с базовой или дополнительном ЛС на квартире. 
* `Дата начала услуги` - дата начала действия услуги на ЛС. 
> * Если на одном ЛС есть несколько базовых услуг (ТО МКД, ТО СВОД или ТО ДВЕРЬ), то ранее действующая должна быть в статусе Завершена. 
***
Структура `bills.xslx`:
* `ЛС` - номер лицевого счета
* `Сумма платежа` - сумма платежа, поступившего на ЛС 
* `Дата загрузки` - дата загрузки платежа в БД
* `Источник платежа` - способ ввода платежа БД. Возможны возвраты платежей со знаком "-" и списания со знаком "+"
* `Способ оплаты` - источник платежа (название платежной системы или оплат в офисе)
* `Комментарий к платежу` - существенная информация по платежу

```python
# Зададим функцию для построения столбчатой диаграммы

def bar_status(parameter, text):
    table = (data[data['service_status'] == parameter]
       .groupby('start_year')
       .agg({'personal_account':'nunique'})
       .sort_values(by='start_year',
                    ascending=False)
       .reset_index())
    table['%'] = table['personal_account'] / table['personal_account'].sum() * 100
    table['%'] = table['%'].round(2)

    return px.bar(table,
                    x='start_year', 
                    y='personal_account',
                    text_auto=True,
                    title=f'Количество {text} услуг ежегодно',
                    text='%')
```

### Доходы и убытки по источнику

#### Доходы

Мы рассчитаем отдельно доходы и расходы компании и посмотрим на соотношения источников между собой.

```python
def revenue_source_g(exclude, text, text_source, parameter='payment_source', barmode='relative'):
    revenue_source = (
           bills[~bills['payment_source'].isin(exclude)].groupby(['upload_month_year', parameter])
           .agg({'payment_sum':'sum'})
           .reset_index()
           .sort_values(by='upload_month_year',
                        ascending=True)
    )
    revenue_source['upload_month_year'] = pd.to_datetime(revenue_source['upload_month_year'],
                                                    format='%Y-%m-%d')
    
    revenue_source['upload_year'] = revenue_source['upload_month_year'].dt.year
    revenue_source['upload_year'] = pd.to_datetime(revenue_source['upload_year'], format='%Y')
    
    revenue_source = (
                revenue_source.groupby(['upload_year', parameter])
                .agg({'payment_sum':'sum'})
                .reset_index()
                .sort_values(by='upload_year', 
                            ascending=True)
                )
    revenue_source['payment_sum'] = revenue_source['payment_sum'].round(2)
    for year in revenue_source['upload_year'].unique():
            revenue_source.loc[revenue_source['upload_year'] == year, ['%']] = revenue_source.loc[revenue_source['upload_year'] == year]['payment_sum'] / revenue_source.loc[revenue_source['upload_year'] == year]['payment_sum'].sum() * 100
    revenue_source['%'] = revenue_source['%'].round(2)
    
    return px.bar(revenue_source,
    x='upload_year',
    y='payment_sum',
    color=parameter,
    title=f'{text} в разбивке по {text_source}: 2018-01-04 по 2021-10-13',
    text=revenue_source['%'].apply(lambda x: '{0:1.2f}%'.format(x)),
    hover_data='%',
    barmode=barmode)
```

### Доходы и убытки по способу оплаты

#### Доходы

Слегка изменим функцию для графиков, добавив туда категоризацию "Другое" для тоносительных величин менее 5%. 

```python
def revenue_method_g(exclude, text, parameter='payment_method', barmode='relative'):
    revenue_source = (
           bills[~bills['payment_source'].isin(exclude)].groupby(['upload_month_year', parameter])
           .agg({'payment_sum':'sum'})
           .reset_index()
           .sort_values(by='upload_month_year',
                        ascending=True)
    )
    revenue_source['upload_month_year'] = pd.to_datetime(revenue_source['upload_month_year'],
                                                    format='%Y-%m-%d')
    
    revenue_source['upload_year'] = revenue_source['upload_month_year'].dt.year
    revenue_source['upload_year'] = pd.to_datetime(revenue_source['upload_year'], format='%Y')
    
    revenue_source = (
                revenue_source.groupby(['upload_year', parameter])
                .agg({'payment_sum':'sum'})
                .reset_index()
                .sort_values(by='upload_year', 
                            ascending=True)
                )
    revenue_source['payment_sum'] = revenue_source['payment_sum'].round(2)
    for year in revenue_source['upload_year'].unique():
            revenue_source.loc[revenue_source['upload_year'] == year, ['%']] = revenue_source.loc[revenue_source['upload_year'] == year]['payment_sum'] / revenue_source.loc[revenue_source['upload_year'] == year]['payment_sum'].sum() * 100
    revenue_source['%'] = revenue_source['%'].round(2)
    revenue_source.loc[revenue_source['%'] < 5, 'payment_method'] = 'Другое'
    
    revenue_source = (
                revenue_source.groupby(['upload_year', 'payment_method'])
                .agg({'payment_sum':'sum'})
                .reset_index()
                .sort_values(by='upload_year', 
                            ascending=True))
    
    revenue_source['payment_sum'] = revenue_source['payment_sum'].round(2)
    for year in revenue_source['upload_year'].unique():
            revenue_source.loc[revenue_source['upload_year'] == year, ['%']] = revenue_source.loc[revenue_source['upload_year'] == year]['payment_sum'] / revenue_source.loc[revenue_source['upload_year'] == year]['payment_sum'].sum() * 100
    revenue_source['%'] = revenue_source['%'].round(2)
    revenue_source = revenue_source.sort_values(by=['upload_year', 'payment_sum'], ascending=False)
    
    return px.bar(revenue_source,
    x='upload_year',
    y='payment_sum',
    color=parameter,
    title=f'{text} в разбивке по способу оплаты: 2018-01-04 по 2021-10-13',
    text=revenue_source['%'].apply(lambda x: '{0:1.2f}%'.format(x)),
    hover_data='%',
    barmode=barmode)
```

### Кумулятивная выручка

```python
df = (plan_rev.merge(pa_payments,
           how='outer', 
           on=['month_year', 'personal_account'],
           sort=True)
      .sort_values(by='month_year')
      .reset_index(drop=True)
     )
df = df.melt(id_vars=['personal_account', 'month_year'], 
             value_vars=['Плановая выручка', 'Сумма платежей'], 
             var_name='category',
             value_name='value')

df['month_year'] = pd.to_datetime(df['month_year'], format='%Y-%m-%d')
df['year'] = df['month_year'].dt.year
df['year'] = pd.to_datetime(df['year'], format='%Y')
df.drop(columns='month_year', axis=1, inplace=False)
df = df.groupby(['month_year', 'category']).agg({'value':'sum'}).reset_index()

for i in df.category.unique():
    df.loc[df['category'] == i, ["value_cum"]] = df.loc[df['category'] == i]["value"].cumsum()

df = df[~df['value_cum'].isna()]
df.loc[df['value'] == 0, ['value_cum']] = np.nan
    
df.head(5)
```

Категория с "Сумма платежей" будет равна 0 вплоть до Января 2018 года, в период, когда у нас доступны данные по платежам.

### Количество оплаченных тарифов 

```python
df = (bills.merge(data[['personal_account', 
                        'service', 
                        'service_rate', 
                        'service_status', 
                        'service_start']],
                  how='left', 
                  on='personal_account',
                  sort=True)
      .sort_values(by=['personal_account', 
                       'upload_date'])
      .reset_index(drop=True))

df.drop_duplicates(inplace=True)

active_df = df[df['service_status'] == 'Активна']

years = [2018, 2019, 2020, 2021]
dictt = {}

for year in years:
    year = year
    start_date = str(year) + "-01-01"
    end_date = str(year) + "-12-31"
    filtered_df = active_df[(active_df['upload_date'] >= start_date) & (active_df['upload_date'] <= end_date)]
    
    grouped_df = filtered_df.groupby('personal_account').agg({'payment_sum':'sum', 'service_rate':'mean'})    
    result_df = grouped_df['payment_sum'] >= grouped_df['service_rate'] * 12    
    
    num_paid_services = result_df.sum()
    
    print(f"Оплаченных тарифов в {year}:", num_paid_services)
    dictt[year] = num_paid_services
```

```python
paid_servs = pd.DataFrame.from_dict(dictt, orient='index', columns=['count'])
paid_servs = paid_servs.reset_index().rename(columns={'index': 'year'})
paid_servs['%'] = paid_servs['count'] / paid_servs['count'].sum() * 100
paid_servs['%'] = paid_servs['%'].round(2)
paid_servs['%_change'] = (paid_servs['count'] / paid_servs['count'].shift(1) - 1) * 100
paid_servs
```

```python
fig = px.bar(paid_servs,
             x='year',
             y='count', 
             title='Количество оплаченных тарифов в год',
             text_auto=True,
             hover_data='%') 
fig.show()
```

6. **Задача:** _Предложить метрики для анализа платежной дисциплины_

Для анализа платежной дисциплины абонентов за услуги можно использовать следующие метрики:

* *Доля просроченных платежей*: это процент платежей, которые не были сделаны в определенный срок.
* *Среднее время просрочки*: это среднее время между датой платежа и датой его просрочки.
* *Процентное соотношение платежей в срок*: это процент платежей, которые были сделаны в определенный срок.
* *Число просроченных платежей на абонента*: это среднее число просроченных платежей на одного абонента.
* *Доля абонентов с просроченными платежами*: это процент абонентов, у которых есть просроченные платежи.

Эти метрики можно использовать в сочетании друг с другом, чтобы получить более полное представление об оплате за услуги. Использование этих метрик позволяет получить более точную картину о том, как абоненты выполняют свои обязательства по оплате за услуги, и поможет в разработке эффективных стратегий для улучшения платежной дисциплины.
